```python
import netCDF4 as nc
import numpy as np
import xarray as xr

# Define input file paths
nc_filen = "/Users/tobias/Downloads/Mipas/MIPAS-E_IMK.200411.V8R_CH4_261.nc"
altout_file = "/Users/tobias/Downloads/Mipas/altout.nc"

import netCDF4 as nc

# Open the original NetCDF file
nc_file = nc.Dataset(nc_filen, mode="r")

# Create a new NetCDF file and copy the data from the original file
altout_file = nc.Dataset("altout.nc", mode="w", format="NETCDF4")
for name, dimension in nc_file.dimensions.items():
    altout_file.createDimension(name, len(dimension) if not dimension.isunlimited() else None)
for name, variable in nc_file.variables.items():
    print(name,variable)
    if name != "altitude":
        out_variable = altout_file.createVariable(name, variable.datatype, variable.dimensions)
        out_variable[:] = variable[:]
        out_variable.setncatts(variable.__dict__)
    else:
        continue

# Close the original and new NetCDF files
nc_file.close()
altout_file.close()


```
